```python
from matplotlib.image import imread
import matplotlib.pyplot as plt
```

```python
image=imread("dashain.png")
plt.figure(figsize=(12,8))
plt.imshow(image)
```

```python
# image=image/255.0
image.shape
```

```python
x=image.reshape(-1,3)
```

```python
x.shape
```

```python
# now performing kmeans clustering
```

```python
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=3).fit(x)
```

```python
img_seg=kmeans.cluster_centers_[kmeans.labels_]
img_seg=img_seg.reshape(image.shape)
```

```python
plt.figure(figsize=(12,10))
```

```python
plt.imshow(img_seg)
```
